```python
import os
from dotenv import load_dotenv
from agno.agent import Agent
from agno.workflow import Workflow
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.models.openai.like import OpenAILike

load_dotenv()

llm_model = OpenAILike(
    id=os.getenv("MODEL_ID", "qwen/qwen3-coder:free"),
    base_url=os.getenv("OPENROUTER_BASE_URL", "https://openrouter.ai/api/v1"),
    api_key=os.getenv("OPENROUTER_API_KEY")
)

web_searcher = Agent(
    name="SearchAgent",
    role="Find information online",
    model=llm_model,
    tools=[DuckDuckGoTools()],
    instructions=[
        "Provide direct answers",
        "Keep responses under 4 sentences",
        "Focus on facts only"
    ]
)

internet_flow = Workflow(
    name="SearchFlow",
    steps=[web_searcher]
)

request = "Кто такая Валя Карнавал(тиктокер)? Какая у неё настоящая фамилия?"

result = internet_flow.run(request, stream=False)

print(result.content)
```

```python
import os
from dotenv import load_dotenv
from agno.agent import Agent
from agno.workflow import Workflow
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.models.openai.like import OpenAILike

load_dotenv()

llm_model = OpenAILike(
    id=os.getenv("MODEL_ID", "qwen/qwen3-coder:free"),
    base_url=os.getenv("OPENROUTER_BASE_URL", "https://openrouter.ai/api/v1"),
    api_key=os.getenv("OPENROUTER_API_KEY")
)

research_agent = Agent(
    name="DetailedResearcher",
    role="Research with comprehensive analysis",
    model=llm_model,
    tools=[DuckDuckGoTools()],
    instructions=[
        "Break down complex queries into steps",
        "Remember previous search results",
        "Build knowledge incrementally",
        "Synthesize information from multiple sources",
        "Provide detailed responses with multiple paragraphs",
        "Include background context and related facts",
        "Add sources and verification details",
        "Explain significance of findings"
    ]
)

planning_flow = Workflow(
    name="DetailedFlow",
    steps=[research_agent]
)

request = "Кто такая Валя Карнавал? Какая у неё настоящая фамилия?"

result = planning_flow.run(request, stream=False)

print(result.content)
```

```python
import os
from dotenv import load_dotenv
from agno.agent import Agent
from agno.workflow import Workflow
from agno.tools.duckduckgo import DuckDuckGoTools
from agno.models.openai.like import OpenAILike

load_dotenv()

llm_model = OpenAILike(
    id=os.getenv("MODEL_ID", "qwen/qwen3-coder:free"),
    base_url=os.getenv("OPENROUTER_BASE_URL", "https://openrouter.ai/api/v1"),
    api_key=os.getenv("OPENROUTER_API_KEY")
)

planner = Agent(
    name="Planner",
    role="Create research strategy",
    model=llm_model,
    instructions=[
        "Break down queries into detailed research phases",
        "Identify key aspects to investigate",
        "Create comprehensive search plan"
    ]
)

researcher = Agent(
    name="Researcher",
    role="Gather comprehensive data",
    model=llm_model,
    tools=[DuckDuckGoTools()],
    instructions=[
        "Search for extensive biographical information",
        "Find career milestones and achievements",
        "Locate personal details and background",
        "Gather multimedia content references",
        "Return detailed structured results"
    ]
)

synthesizer = Agent(
    name="Synthesizer",
    role="Analyze and elaborate",
    model=llm_model,
    instructions=[
        "Combine all gathered information",
        "Create multi-paragraph comprehensive response",
        "Include background, career, and personal details",
        "Add context about related projects and shows",
        "Provide detailed explanation with multiple sections"
    ]
)

multi_agent_flow = Workflow(
    name="ExtendedMultiAgentFlow",
    steps=[planner, researcher, synthesizer]
)

request = "Кто такая Валя Карнавал? Какая у неё настоящая фамилия? Давай подробную биографию"

result = multi_agent_flow.run(request, stream=False)

print(result.content)
```
