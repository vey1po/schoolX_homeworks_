```python
class CanSwim:
    def swim(self):
        print(f"{self.name} is swimming!")

class CanDance:
    def dance(self):
        print(f"{self.name} is dancing!")

class CanFly:
    def fly(self):
        print(f"{self.name} is flying!")

class CanTalk:
    def talk(self):
        print(f"{self.name} is talking!")

class CanRun:
    def run(self):
        print(f"{self.name} is running fast!")

class CanFight:
    def fight(self):
        print(f"{self.name} is fighting!")

class CanCook:
    def cook(self):
        print(f"{self.name} is cooking delicious food!")


class BaseCharacter:
    def __init__(self, name):
        self.name = name
    
    def greet(self):
        print(f"Hi, my name is {self.name}")

class BaseInActionCharacter(BaseCharacter):
    def __init__(self, name, action_level):
        super().__init__(name)
        self.action_level = action_level

class BaseFunkoPop(BaseCharacter):
    def __init__(self, name, material):
        super().__init__(name)
        self.material = material
    
    def display(self):
        print(f"Funko Pop of {self.name} made of {self.material}")

class BaseCosplayer(BaseCharacter):
    def __init__(self, name, skill_level):
        super().__init__(name)
        self.skill_level = skill_level
    
    def perform(self):
        print(f"Cosplayer {self.name} performs with skill level {self.skill_level}")

class BaseHuman(BaseCharacter):
    def __init__(self, name, age):
        super().__init__(name)
        self.age = age
    
    def introduce(self):
        print(f"I'm {self.name}, {self.age} years old")


class Shrek(BaseInActionCharacter, CanSwim, CanTalk, CanCook):
    def __init__(self):
        super().__init__("Shrek", "high")
    
    def greet(self):
        print("Ogre power! I'm Shrek!")

class PussInBoots(BaseInActionCharacter, CanTalk, CanFight, CanRun):
    def __init__(self):
        super().__init__("Puss In Boots", "very high")
    
    def greet(self):
        print("Shall we dance? I'm Puss In Boots!")

class Donkey(BaseInActionCharacter, CanTalk, CanDance, CanRun):
    def __init__(self):
        super().__init__("Donkey", "medium")
    
    def greet(self):
        print("I talk, therefore I am! Hi, I'm Donkey!")

class JackHorner(BaseHuman, CanTalk, CanFly):
    def __init__(self):
        super().__init__("Jack Horner", 45)
    
    def greet(self):
        print("I'm Jack Horner, and I can fly!")

# Фанко Попы
class ShrekFunkoPop(BaseFunkoPop, CanSwim):
    def __init__(self):
        super().__init__("Shrek", "vinyl")
    
    def display(self):
        print(f"Shrek Funko Pop made of {self.material}")
        self.swim()

class PussInBootsFunkoPop(BaseFunkoPop, CanFight):
    def __init__(self):
        super().__init__("Puss In Boots", "vinyl")
    
    def display(self):
        print(f"Puss In Boots Funko Pop made of {self.material}")
        self.fight()

class DonkeyFunkoPop(BaseFunkoPop, CanDance):
    def __init__(self):
        super().__init__("Donkey", "vinyl")
    
    def display(self):
        print(f"Donkey Funko Pop made of {self.material}")
        self.dance()

class JackHornerFunkoPop(BaseFunkoPop, CanFly):
    def __init__(self):
        super().__init__("Jack Horner", "vinyl")
    
    def display(self):
        print(f"Jack Horner Funko Pop made of {self.material}")
        self.fly()

# Косплееры
class ShrekCosplayer(BaseCosplayer, CanCook, CanSwim):
    def __init__(self):
        super().__init__("Shrek", "expert")
    
    def perform(self):
        print(f"Cosplayer {self.name} with skill {self.skill_level}")
        self.cook()
        self.swim()

class PussInBootsCosplayer(BaseCosplayer, CanFight, CanRun):
    def __init__(self):
        super().__init__("Puss In Boots", "master")
    
    def perform(self):
        print(f"Cosplayer {self.name} with skill {self.skill_level}")
        self.fight()
        self.run()

class DonkeyCosplayer(BaseCosplayer, CanDance, CanTalk):
    def __init__(self):
        super().__init__("Donkey", "intermediate")
    
    def perform(self):
        print(f"Cosplayer {self.name} with skill {self.skill_level}")
        self.dance()
        self.talk()

class JackHornerCosplayer(BaseCosplayer, CanFly, CanTalk):
    def __init__(self):
        super().__init__("Jack Horner", "advanced")
    
    def perform(self):
        print(f"Cosplayer {self.name} with skill {self.skill_level}")
        self.fly()
        self.talk()

shrek = Shrek()
puss = PussInBoots()
donkey = Donkey()
jack = JackHorner()

shrek_pop = ShrekFunkoPop()
puss_pop = PussInBootsFunkoPop()
donkey_pop = DonkeyFunkoPop()
jack_pop = JackHornerFunkoPop()

shrek_cos = ShrekCosplayer()
puss_cos = PussInBootsCosplayer()
donkey_cos = DonkeyCosplayer()
jack_cos = JackHornerCosplayer()

print("Characters")
shrek.greet()
shrek.swim()
shrek.talk()
shrek.cook()

puss.greet()
puss.talk()
puss.fight()
puss.run()

donkey.greet()
donkey.talk()
donkey.dance()
donkey.run()

jack.greet()
jack.talk()
jack.fly()
jack.introduce()

print("Funko Pops")
shrek_pop.display()
puss_pop.display()
donkey_pop.display()
jack_pop.display()

print("Cosplayers")
shrek_cos.perform()
puss_cos.perform()
donkey_cos.perform()
jack_cos.perform()
```
